```python
campaigns = [
    {"name": "검색광고", "cost": 100000, "sales": 350000},
    {"name": "SNS광고", "cost": 200000, "sales": 180000},
    {"name": "배너광고", "cost": 150000, "sales": 600000}
]

def print_campaign(ad):
    print("광고명:", ad["name"])
    print("광고비:", ad["cost"])
    print("매출:", ad["sales"])
    print("-----")

for campaign in campaigns:
    print_campaign(campaign)
```
