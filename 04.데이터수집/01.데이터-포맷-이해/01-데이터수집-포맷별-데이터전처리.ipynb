```python
%%writefile sample.json

[
    {
      "id": "A001",
      "city": "Tokyo",
      "country": "Japan",
    },
    {
      "id": "B012",
      "city": "Jerusalem",
      "country": "Isreal",
    },
    {
      "id": "C123",
      "city": "Busan",
      "country": "South Korea",
    }
]
```

```python
# JSON은 JavaScript Object Notation의 약자로,
# 데이터를 구조화하여 저장하거나 전송할 때 사용하는 포맷
# 주로 웹에서 클라이언트와 서버 간 데이터 교환에 사용

# 가볍고 텍스트 기반이라 사람이 읽고 쓰기 쉬움
# 대부분의 프로그래밍 언어에서 쉽게 읽고 생성 가능
# **키-값 쌍(key-value pair)**으로 구성된 데이터 구조 사용

df_json = pd.read_json("sample.json")

df_json.info()
df_json.describe()
```

```python
# 데이타 프레임을 json 으로 변환 및 저장
import pandas as pd

data = {
        '이름': ['John', 'Doe', 'Jane'],
        '나이': [30, 25, 28],
        '성별': ['남성', '남성', '여성']
        }

df = pd.DataFrame(data)

print(df.head())

#데이타프레임을 json 으로 바꿈
json_data = df.to_json(orient="records",indent=4)
print(json_data)

# 데이타 프레임을 JSON 파일로 바로 저장
# 각 행을 JSON 객체로 변환 (리스트 형태)
# df.to_json('행기준.json', orient="records",force_ascii=False,indent=4)
# df.to_json('열기준.json', orient="columns",force_ascii=False,indent=4)
```
